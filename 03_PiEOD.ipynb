#Phyton'da İf-Else Koşulları
Phyton'da değişkenlerin içindeki değerlere göre işlem yapabilmemizi sağlar. Programlarımızda kullanıcıdan aldığımız değerlere göre koşullar oluşturup kullanıcının isteğine göre işlemler yapabilirim.

````python
#İf-Else Örnek Kullanımı
#Boolean Değişkenş sadece True (Doğru) veya False (Yanlış) değeri alan değişkenlerdi
cinsiyet = True #Kadınsa True erkekse  False olsun.
if cinsiyet == True: #Eğer cinsiyet kadınsa...
  print("Merhaba hanımefendi")
else: #değilse (kadın değilse)
  #İf veya else içinde tab ile boşluk bırakarak yazarım.
  print("Merhaba beyefendi")
print("İyi günler^^")
````

````python
print("Merhaba")
sayi1 = int(input("Lütfen üçgenin taban uzunluğunu giriniz: "))
sayi2 = int(input("lütfen üçgenin uzunluğu giriniz:"))
print(f"Üçgenin alanı = {sayi1*sayi2/2}")
````

````python
sayi1 = int(input("Dairenin yarı çapını giriniz: "))
print(f"Dairenin alanı = {3.14*sayi1*sayi1}")
````

##Önemli Noktalar
---


1. **MANTIKSAL KOŞULLAR**

Eğer değilse gibi yapılarda kullanılan karşılaştırmayı sağlayan mantıksal operatörler olarak bilinmektedir. Bu mantıksal operatörler değişkenlerin içindeki değerlerin eşitolup olmadığını, büyük veya küçük olup olmamasına göre karar vermenizi sağlar.

**İŞARETLER**

**(==)** -> Eşittir.

**(!=)** -> Eşit Değildir.

**(>)**  -> Büyüktür.

**(<)**  -> Küçüktür.

**(>=)** -> Büyük ve Eşittir.

**(<=)** -> Küçük ve Eşittir.

**Not:** İşaretler arasında boşluk bırakmayın (== veya != gibi yapın.)

2. **İKİ NOKTA ÜST ÜSTENİN ANLAMI**

İki nokta üst üste Phyton'da altındaki satırın üstteki satırın önemli komutları olduğunu göstermeketdir.

**ÖRNEĞİN**

Koşul:
  Koşul sağlandığınında çalışan kod
```
if sayi == 1:
  print("sayinin degeri 1'dir")
```
**DİKKAT:**  İki nokta üst üste koyulduktan sonra alttaki satır dört boşlukla **yazılıyor**.


````python
cinsiyet = input("cinsiyetinizi lütfen giriniz")
if cinsiyet == 'erkek' or cinsiyet != 'kadın' or cinsiyet == 'ERKEK':
    print("Merhaba Beyfendi")

elif cinsiyet == 'kadın' or cinsiyet != 'erkek' or cinsiyet == 'KADIN':
    print("Merhaba Hanımefendi")

else:
    print("Lütfen cinsiyetinizi doğru giriniz")
````

````python
Not = int(input("Sınavdan aldığınız puan "))
if Not >= 85 and Not <= 100:
  print("Sınavı pek iyi ile geçtiniz")
elif Not >= 70 and Not <= 84:
  print("Sınavı iyi ile geçtiniz")
elif Not >= 50 and Not <= 69:
  print("Sınavı orta ile geçtiniz")
elif Not >= 30 and Not <= 49:
  print("Sınavı düşük ile geçtiniz")
elif Not >= 0 and Not <= 29:
    print("Sınavdan kaldınız")
else:
  print("Lütfen bir not giriniz")
````
