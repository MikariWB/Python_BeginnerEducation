#While Cycle
While cycle, unlike for, it works if a condition is met. This condition also works on logical conditions other than numbers. It is basically a mixture of if and for. For example;

1. If the value of a is less than 40, work as long as,
2. If password is TRUE, it will work as long as it is
3. If username is "Mikari" it will work as long as it is

While always needs a condition.

```python
while(True):
  print("Mikari^^")
```

Note: The above example runs forever. It is harmful to computer health.

```python
number = 0
while(number <= 10): #Number is equal to 10 and smaller then 10, keep on working
  print(number)
  number += 1 #Add numbers to the number one by one.
```

```python
information = input("Write your username: ")
while(information != "Mikari^^"):
  print("Wrong username")
  information = input("Write your username: ")
print("Correct username, welcome Mrs Mikari")
```

```python
information = input("Please enter a letter from your keyboard: ")
while (information != "x"):
  print("The value you enter is: " + information)
  information = input("Please enter a letter from your keyboard: ")








```
