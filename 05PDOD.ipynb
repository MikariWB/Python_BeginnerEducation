#Loops (For and While Loop)

---
Let's say you have multiple variables or a lot of repetitive processes. That's when we use loops. For example;

1. If I want to print numbers from 1 to 1000,
2. I have a list of 1000 variables and I want to print it,
3. If I have processes that will run as long as a desired condition is met,
4. If I have repeating loops, I use loops.

There are two different cycles in total.  One is a **For** loop and the other is a **While** loop.

Now let's examine the **For Loop**.


```python
#What needs to be done to set up the for loop is quite simple:
for i in range(1,10):
  print(i)
#It doesn't always get the last line so I write one digit more.
#I write 101, not 100.
#Let's examine the loop above:
#for ... in ... : is a pattern.
#What says i is a variable.
#range means a range.
#Let's do another example:
name = "Mikari"
for a in name:
  print(a)
#I can break down the letters of my name with a for loop.
#I can print line by line.
```

**In the for loop, range() can take three values. The first is the number it will start, the second is the number it will end, and the third is the number by which the cycle will increase.**

```python
for a in range(0,50):
  if (a % 5 == 0):
    print(a)
  else:
    print("The number is not divisible by 5, thats not the number!")
```


**Note:** The **% (Percent)** sign means taking the mode of the number. For example, the remainder after dividing a number. For example, the remainder of the number "a" divided by 6 is called mod.:


```python
for i in range(0,50):
  if(i % 15 == 0):
    print("Number divisible by 15: ",i)
  if(i % 5 == 0 or i % 3 == 0):
    print(i)
```

# Can Have Nested For Loops
Sometimes more than one for loop can be written inside each other. This is a bit complicated. It takes a lot of practice.

```python
for a in range (0,5):
  print("variable a wrote: ",a)
  for b in range(0,5):
    print("variable b wrote: ",b)
```

```python
number1 = int(input("Please write a number: "))
number2 = int(input("Please write another number: "))
for a in range(number1,number2):
  if a % 2 == 0 or a % 3 == 0:
    print(a)
```

```python
for a in range(1,50):
  for b in range(1,100):
    print(a,"the loop value b returned in: ",b)
```

```python
for a in range(1,10):
  print(a,"value")
  for b in range(1,50):
    if b % 15 == 0:
      print(b," values inside ",a)


```

```python
#Adding numbers in a for loop
total = 0
information = int(input("Please write the number you will add from 1 to the number you will add:"))
for number in range(1,information+1):
  total += number #total = total + number
print("value of total variable: ",total,"added number is: ",number)
```
