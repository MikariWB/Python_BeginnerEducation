# Different operations on lists
We can perform operations such as clearing lists, selecting elements and removing them at the same time, printing the list in reverse, and getting list element count or index information.

```python
# Clear Command -> Delete everything.
day_of_week = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
day_of_week.clear()
print(day_of_week)
```

```python
# Count Command -> Number of written elements
day_of_week = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
print(day_of_week.count("Monday"))
```

```python
# Index Command -> Writes the index of the written element.
day_of_week = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
print(day_of_week.index("Monday"))
```

```python
# Pop Command -> Takes the indexed elements from the list, writes them and removes them from the list.
day_of_week = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
day = day_of_week.pop(2)
print(day_of_week.pop(2))
print(day)
print(day_of_week)
```

```python
# Reverse Command -> Reverse the list
day_of_week = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
day_of_week.reverse()
print(day_of_week)
```

# Searching Lists and Using Loops
In order to get information that will be printed or useful when certain conditions are met in the lists, searches must be done in the lists.

Loops can be used to search in lists.

```python
# Printing numbers between 0 and 250 by 3
by_three_list = []
multiple_of_fifteen = []
for a in range(0,250,3):
    by_three_list.append(a)
print(by_three_list)
print("total number of elements",len(by_three_list))
#
for element in range(0, len(by_three_list)):
    if by_three_list[element] % 5 == 0:
        multiple_of_fifteen.append(by_three_list[element])
print(multiple_of_fifteen)
print("total number of elements",len(multiple_of_fifteen))
```
