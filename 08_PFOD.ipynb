# Functions - Methods
Functions, known as methods in different programming languages, are known as "Object Oriented Programming (OOP)". This programming (OOP) aims to store operations and variables that can be used repeatedly or later in the software.

**For example**

1. Let's say you want to use a list more than once. OOP allows us to write it once and call it again in the following codes.
2. Let's say we are doing a mathematical operation. Hiding this operation inside a function and calling that function prevents us from writing lines of code.
3. We can use the operations we wrote in a program by adding them to another program or adding them to its files. We do this with import. We can create our own libraries just like adding libraries.

#GPA calculation

Lets say, there are 3 exams in math class. If we want to calculate GPA of this class, we  add the 3 exam grades and divide by 3. Same in the Turkish lesson, there 4 exams. We add the 4 exam grades and divide by 4.



```
conclusion: I add up the exam grades of the number of exams I take and divide them by the number of grades.

```

````python
# 1.method
def GPA():
  number = int(input("Enter your note(write -1 to leave):"))
  addition = 0
  GPA = 0
  entered_grades = 1
  while number != -1:
    addition += number
    entered_grades += 1
    number = int(input("Enter your note(write -1 to leave):"))
    if(number == -1):
      break
    GPA = addition / entered_grades
  print("GPA is", GPA)
  if GPA >= 85 and GPA <= 100:
    print("You passed the test with very good")
  elif GPA >= 70 and GPA <= 84:
    print("You passed the test with great")
  elif GPA >= 50 and GPA <= 69:
    print("You passed the test with average")
  elif GPA >= 30 and GPA <= 49:
    print("You passed the test with Bad")
  elif GPA >= 0 and GPA <= 29:
    print("You didn't pass the test")
  else:
     print("Please enter a grade")


````

````python
GPA()
````

````python
# 2.method - With For
def GPA_F(number_of_numbers):
  addition = 0
  GPA = 0
  for a in range(0,number_of_numbers):
    numbers = int(input("Please enter your grade: "))
    addition += numbers
  GPA = addition / number_of_numbers
  print("GPA is", GPA)
````

````python
number = int(input("Please write how many grades you gonna enter: "))
GPA_F(number)
````
