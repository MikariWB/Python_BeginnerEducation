#Değişkenlerle Dört İşlem
Toplama, çıkarma, çarpma ve bölme işlemleri kolaylıkla yapılabilir.

Sayılar int yani integer değerlerle yazılabilir. Tamsayı anlamına gelir.

```python
sayi_1 = int(input("Birinci sayıyı giriniz: "))
sayi_2 = int(input("İkinci sayıyı giriniz: "))
#Dört İşlemi ismine göre değişkenelerde yazalım.
toplam = sayi_1 + sayi_2
cikarma = sayi_1 - sayi_2
carpma = sayi_1 * sayi_2
bolme = sayi_1 / sayi_2
print(f"Toplama:{toplam}, Çıkarma:{cikarma}, Çarpma:{carpma}, Bölme:{bolme}")
```

```python
#Bir karenin ve bir dikdörtgen alanını hesaplayınız.
#Karenin alanı: kenar*kenar=alan.
#Dikdörtgenin alanı: kkenar*ukenar=alan.
kare_k = (input("Karenin kenarını giriniz: "))
kare_k = int(kare_k)
kare_alan = kare_k * kare_k
print("Karenin alanı:", kare_alan)
dikdörtgen_kk = (input("Dikdörtgenin kısa kenarını giriniz: "))
dikdörtgen_kk = int(dikdörtgen_kk)
dikdörtgen_uk = (input("Dikdörtgenin uzun kenarını giriniz: "))
dikdörtgen_uk = int(dikdörtgen_uk)
dikdörtgen_alan = dikdörtgen_kk * dikdörtgen_uk
print("Dikdörtgenin alanı:", dikdörtgen_alan)
```
